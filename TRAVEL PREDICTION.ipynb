```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing and reading json file
import json 
with open('challenge_3.json', 'r') as f:
    distros_dict = json.load(f)
distros_dict
```

```python
#flattening a nested json file and converting json to pandas datafram
from pandas.io.json import json_normalize
from flatten_json import flatten
dic_flattened = [flatten(d) for d in distros_dict]
df=pd.DataFrame(dic_flattened)
df.columns=['cities','session_id','unix-timestamp','user_row','country','user_joining_date','user_id']
df=df[['user_id','session_id','country','cities','unix-timestamp','user_joining_date']]
df.head(100)
```

```python
#checking if any null value is there
null_data = df.columns[df.isnull().any()]
df[null_data].isnull().sum()
```

```python
df['country'].isnull;
```

```python
#grouping by data to find the missing country 
df.groupby('country')['user_id'].count()
```

```python
df['cities'];
```

```python
#splitting cities by comma to have first and second priorities of cities seperated
df1=df['cities'].str.split(',',expand=True)
df1.head()
```

```python
#filling missing values by 0
df1=df1.fillna(0)
df1.head()
```

```python
# replacing all zero values in the second city section to first city name in the same row 
for i in range(len(df1)):
    if df1.loc[i,1]==0:
        df1.loc[i,1]=df1.loc[i,0]

df1.head()
```

```python
#determining first and second sets of cities
df_first_city=df1.loc[:,0]
df_sec_city=df1.loc[:,1]
df1.loc[:,1];
```

```python
# changing all city names to numerical values
from sklearn import preprocessing
encoding_cities = preprocessing.LabelEncoder()
 
cities_coded=encoding_cities.fit_transform(df1.loc[:,1])

target_data=pd.DataFrame(cities_coded,columns=['second_city'])
```

```python
# encoding first cities to make them ready for training 
import category_encoders as ce
ohe = ce.OneHotEncoder(handle_unknown='ignore', use_cat_names=True)
training_data=ohe.fit_transform(df1.loc[:,0])
```

```python
training_data;
```

```python
#importing libraries for training data (linear regression), splitting data and chacking the accuracy of model
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(training_data, target_data, test_size=0.2)
```

```python
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# r squared to show the performance of model
r2=r2_score(y_test,y_pred)
print('R2 score: ', r2)
```

```python
#importing logistic regression
from sklearn.linear_model import LogisticRegression
logisticreg = LogisticRegression()
logisticreg.fit(X_train, y_train)
y_pred_log = logisticreg.predict(X_test)
logisticreg.score(X_test, y_test)
```

```python
likelihood = logisticreg.predict(training_data)
likelihood=pd.DataFrame(likelihood)
data_most_likelihood_city=pd.concat([df_first_city,likelihood,df_sec_city,target_data],axis=1)
data_most_likelihood_city.columns=["actual_first_searched_city","(predicted_second_city)most likely city as second city","actual_second_searched_city","Numbers corresponding to second cities"]
data_most_likelihood_city.to_csv('data_most_likelihood_city')
data_most_likelihood_city
```

```python
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
random_model = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
random_model.fit(X_train,y_train)
predictions=random_model.predict(X_test)
random_model.score(X_test, y_test)
```

```python
distance_data=distance_dataframe['distance_value']
```

```python
# rearranging first, second cities and related distance 
final_df=df1.loc[:,0:1]
final_df.columns=['first_city','second_city']
final_df['distance']=distance_data
```

```python
final_df
```

```python
#calculating the mean of distances 
mean_distance=final_df['distance'].mean()
mean_distance=round(mean_distance)
```

```python
# grouping customers based on the distance between cities
grouped_customers=[]
for i in range(len(final_df)):
    if final_df['distance'][i]<mean_distance:
        cat="high intent"
        grouped_customers.append(cat)
    else:
        cat="low intent"
        grouped_customers.append(cat)
```
